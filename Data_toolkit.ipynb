3/ Explain the following terms:
The difference in np.array, np.asarray and np.asanyarrayX
The difference between Deep copy and shallow copy

-
np.array: This function always creates a new array, even if you pass an existing NumPy array to it. It essentially copies the data to a new array.

np.asarray: This function is similar to np.array, but it's more flexible. If you pass an existing NumPy array to np.asarray, it might return the original array without creating a copy if the data type and order are compatible. Otherwise, it creates a new copy.

np.asanyarray: This function is the most lenient of the three. It's designed to work with array-like objects, including subclasses of ndarray. It will return the object as a NumPy array if possible, without creating a copy unless necessary.

-
Shallow Copy: A shallow copy creates a new object but doesn't create copies of the objects contained within it. Instead, it creates references to the original objects. This means that if you modify an element in the original object, the change will also be reflected in the shallow copy, and vice versa.

Deep Copy: A deep copy creates a completely independent copy of the original object and all the objects it contains. Modifying elements in the deep copy won't affect the original object, and vice versa.

14/ Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample time-series data
num_days = 30
dates = pd.date_range(start='2023-10-27', periods=num_days)
temperatures = np.random.randint(20, 35, size=num_days)  # Sample temperatures
humidity = np.random.randint(40, 90, size=num_days)     # Sample humidity

# Create DataFrame
data = pd.DataFrame({'Date': dates, 'Temperature': temperatures, 'Humidity': humidity})
data.set_index('Date', inplace=True)  # Set 'Date' as index for time-series

# Create the plot with two y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(data.index, data['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='red')
ax1.tick_params('y', labelcolor='red')

# Create a second y-axis for Humidity on the right
ax2 = ax1.twinx()
ax2.plot(data.index, data['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params('y', labelcolor='blue')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Display legend
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=2)
# Adjust legend position

# Show the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

15/ Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:
a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate data from a normal distribution
data = np.random.normal(size=1000)  # Mean=0, Std Dev=1 by default

# a) Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 0, 1)  # Mean=0, Std Dev=1
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# c) Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# d) Display legend
plt.legend()

# Show plot
plt.show()
```

17/ Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data for x and y
x = np.random.rand(50) - 0.5  # Center around 0
y = np.random.rand(50) - 0.5  # Center around 0

# Create a DataFrame for Seaborn
import pandas as pd
df = pd.DataFrame({'x': x, 'y': y})

# Define a function to assign quadrants
def assign_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant 1'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant 2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

# Apply the function to create a 'Quadrant' column
df['Quadrant'] = df.apply(assign_quadrant, axis=1)

# Create the scatter plot using Seaborn
sns.scatterplot(x='x', y='y', hue='Quadrant', data=df, palette='viridis')
# 'palette' for color scheme

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
